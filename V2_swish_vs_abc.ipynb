```python
import re 

def extract_flag_id(details):
    match = re.search(r'ABC_FLAG_IDS=\[(\d+)\]', details)
    return match.group(1) if match else None 
```

```python
swish['details'].apply(extract_flag_id)
```

```python
#Adding flag_id to swish
swish['FLAG_ID'] = swish['details'].apply(extract_flag_id)
swish.head()
```

```python
#renaming some of the columns 
swish = swish.rename(columns = {
    'feed_value': 'feed_valu_swish',
    'details': 'details_swish'
})
swish.head()
```
